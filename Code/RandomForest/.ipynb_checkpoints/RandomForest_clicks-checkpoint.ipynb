```python
#Load data
full_data = pd.read_csv('../clean_data/pix_full_data_with_topics_features.csv')
```

# Model 1: (almost) all features

```python
X = full_data.ix[:,['word_count','ALLCAPS','topic1','topic2','topic3','topic4','topic5','topic6','topic7','topic8','video','sex','number']]

#,'year','city','women','politics' are other features

hour = pd.get_dummies(full_data.hour)
day = pd.get_dummies(full_data.day)
monthyear = pd.get_dummies(full_data.monthyear)
Predictor = pd.concat([hour,day,monthyear,X],axis=1)
#X['day'] = X['day'].astype('category')
#vect = DictVectorizer()
#X = vect.fit_transform(X).toarray()

Response = full_data.ix[:,'link_buckets']
```

```python
Predictor_train, Predictor_test, Response_train, Response_test = train_test_split(Predictor, Response, random_state=0)
clf = RandomForestClassifier(n_estimators=100, random_state=10).fit(Predictor_train, Response_train)
Response_prob = clf.predict_proba(Predictor_test)[:,1]

```

```python
def threshold(x):
    if x>0.46:
        return 1
    else:
        return 0
threshold = np.vectorize(threshold)
Response_pred = threshold(Response_prob)
accuracy = metrics.accuracy_score(Response_test, Response_pred)
precision = metrics.precision_score(Response_test, Response_pred)
recall = metrics.recall_score(Response_test, Response_pred, average='micro')
f1_score = metrics.f1_score(Response_test, Response_pred, average='micro')
```

These are ways of taking into account not just the classification results, but the results **relative to the true category**.

$$ {\rm accuracy} \equiv \frac{\rm correct~labels}{\rm total~samples} = 0.673$$

$$ {\rm precision} \equiv \frac{\rm true~positives}{\rm true~positives + false~positives} = 0.638$$

$$ {\rm recall} \equiv \frac{\rm true~positives}{\rm true~positives + false~negatives} = 0.783$$

$$ F_1 \equiv 2 \frac{\rm precision \cdot recall}{\rm precision + recall} $$

The **accuracy**, **precision**, **recall**, and **f1-score** all range from 0 to 1, with 1 being optimal.
Here we've used the following definitions:

- *True Positives* are those which are labeled ``1`` which are actually ``1``
- *False Positives* are those which are labeled ``1`` which are actually ``0``
- *True Negatives* are those which are labeled ``0`` which are actually ``0``
- *False Negatives* are those which are labeled ``0`` which are actually ``1``

```python
confusion_matrix(Response_test,Response_pred)
```

```python
%matplotlib inline
fig, ax = plt.subplots(1)
ax.plot([0, 1], [0, 1], 'k--', label='Random classifier')
fpr, tpr, _thresholds = metrics.roc_curve(Response_test, Response_prob)
auc = metrics.auc(fpr, tpr)
ax.plot(fpr, tpr, label='All features AUC = %0.2f' % auc)

ax.set_xlabel('False Positive Rate', fontsize=14)
ax.set_ylabel('True Positive Rate', fontsize=14)
ax.set_title('Receiver Operator Characteristic', fontsize=16)
ax.legend(loc='lower right', fontsize=14)
# fig.savefig('basic_roc.png')
```
